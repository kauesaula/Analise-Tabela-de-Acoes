```python
# Importando as bibliotecas a serem utilizadas
import pandas as pd
import plotly.express as px
```

```python
# Definindo um dataframe para a aba "Principal" da planilha
# Note que é utilizada a função .head(), que, neste caso, foi usada para mostrar apenas as 10 primeiras linhas da tabela
df_principal = pd.read_excel('/content/drive/MyDrive/Python/Imersão Python/tabela_acoes.xlsx', sheet_name='Principal')
df_principal.head(10)
```

```python
# Definindo um dataframe para a aba "Total_de_acoes" da planilha
df_total_acoes = pd.read_excel('/content/drive/MyDrive/Python/Imersão Python/tabela_acoes.xlsx', sheet_name='Total_de_acoes')
df_total_acoes
```

```python
# Definindo um dataframe para a aba "Ticker" da planilha
df_ticker = pd.read_excel('/content/drive/MyDrive/Python/Imersão Python/tabela_acoes.xlsx', sheet_name='Ticker')
df_ticker
```

```python
# Definindo um dataframe para a aba "Chatgpt" da planilha
  # O ChatGPT foi utilizado para adquirir informações referentes ao segmento e à "idade" das empresas
df_gpt = pd.read_excel('/content/drive/MyDrive/Python/Imersão Python/tabela_acoes.xlsx', sheet_name='Chatgpt')
df_gpt
```

```python
# Filtrando o dataframe para que ele mostre apenas as opções desejadas
# A função .copy() basicamente faz uma cópia do dataframe que já existe apenas com as colunas incluídas na linha do código e armazena esta “nova informação” no dataframe.
  # Caso o comando não seja inserido, a linha fica atrelada ao dataframe original, porém, como fora inserido, o comando original é substituído pelo novo.

df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
# Renomeando as colunas com caractéres especiais, visando evitar qualquer tipo de infortúnio na execução do código
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
# Realizando os cálculos para que possamos incluir novas colunas, sendo elas a coluna "Var_pct" e a coluna "valor_inicial"
  # Para criar a coluna "Var_pct", foi necessário dividir o valor da coluna "var_dia_pct" por 100
  # Para criar a coluna "valor_inicial", foi dividido o valor da coluna "valor_final" ao valor da coluna "Var_pct" somado a 1
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
# Conectando as abas "Principal" e "Total_de_acoes" com a função .merge()
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
# Excluindo a coluna "Código", visto que é desnecessária
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
# Novamente renomeando colunas
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtde_teorica'}).copy()
df_principal
```

```python
# Criando uma nova coluna referente à variação em reais (R$) das ações
# Para isso, multiplicou-se o valor da subtração entre a coluna "valor_final" e "valor_inicial" pelo valor da coluna "qtde_teorica"
df_principal['var_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['qtde_teorica']
df_principal
```

```python
# Formatando todos os valores do tipo float presentes no notebook, a fim de evitar problemas com a leitura de tais valores.
pd.options.display.float_format = '{:.2f}'.format
df_principal
```

```python
# Registrando a coluna "qtde_teorica" como portadora de dados do tipo inteiro
df_principal['qtde_teorica'] = df_principal['qtde_teorica'].astype(int)
df_principal
```

```python
# Definindo a coluna "Resultado", utilizando a função .apply() para indicar caso o resultado tenha subido, descido ou esteja estável.
df_principal['Resultado'] = df_principal['var_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
# Definindo a coluna referente ao nome das empresas mesclando a aba "Principal" e a aba "Ticker"
# Em seguida, dropando a coluna "Ticker" da planilha (aba e coluna compartilham do mesmo nome, não confundir)
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
# Repetindo o mesmo processo, porém adicionando a aba "Chatgpt" à tabela principal e dropando a coluna que possui o nome das empresas, visto que ela já está presente na tabela principal
df_principal = df_principal.merge(df_gpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
# Renomeando mais um conjunto de colunas por fins estéticos
df_principal = df_principal.rename(columns={'Data':'data', 'Resultado':'resultado', 'Nome':'nome', 'Segmento':'segmento', 'Idade (anos)':'idade'}).copy()
df_principal
```

```python
# Definindo um dataframe para a aba "Análises" da planilha
df_analise = pd.read_excel('/content/drive/MyDrive/Python/Imersão Python/tabela_acoes.xlsx', sheet_name='Análises')
df_analise
```

```python
# Definindo no código os valores contidos na aba de Análises
maior = df_principal['var_rs'].max()
menor = df_principal['var_rs'].min()
media = df_principal['var_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['var_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['var_rs'].mean()
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
# Criandp um dataframe referente apenas aos valores que subiram
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
# Utilizando a função .groupby() para somar tudo por segmento
df_analise_segmento = df_principal_subiu.groupby('segmento')['var_rs'].sum().reset_index()
df_analise_segmento
```

```python
# Repetindo o processo para podermos analisar o saldo
df_analise_saldo = df_principal.groupby('resultado')['var_rs'].sum().reset_index()
df_analise_saldo
```

```python
# Por fim, plotando um gráfico que nos mostrará a variação em reais por resultado
fig = px.bar(df_analise_saldo, x='resultado', y='var_rs', text='var_rs', title='Variação em Reais por Resultado')
fig.show()
```
